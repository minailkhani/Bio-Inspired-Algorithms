show stocks and orders in them:

```python
def show(order, stockLen):
    usedSpace = 0
    s = []
    for element in order:
        usedSpace+=element
        s.append(element)
        if usedSpace>stockLen:
            print(s[:-1], "usedSpace:", usedSpace-element)
            usedSpace = element
            s = [element]
    if s != []:
        print(s, "usedSpace:", usedSpace)
    
```

cost is the number of used stocks

```python
def cal_cost(order,stockLength):
    n = 1
    usedSpace = 0
    for element in order:
        usedSpace+=element
        if usedSpace>stockLength:
            n+=1
            usedSpace = element
    return n
```

create neighbours:

r is radius of neighborhood & n is num of Neighbors

each neighbour created by swapping r random order

we can choose the neighbour or not. it can be ranomly or choosing the best one

```python
def select_neighbour(stocks,numOfOrders, len):
    neighbours = []
    global n
    for k in range(n):
        neighbours.append(stocks.copy())
        index = random.sample(range(numOfOrders), k=r)
        order = random.sample(range(r),k=r)
        for i in range(r):
            neighbours[k][index[i]] = stocks[index[order[i]]]
    if not rand:
        costs = []
        for ne in neighbours:
            newCost = cal_cost(ne, len)
            costs.append(newCost)
        return neighbours[costs.index(max(costs))], newCost 
    else:
        ans = random.choice(neighbours)
        return ans,   cal_cost(ans, len)
```

T tend to 0

simulated annealing! clear :)

```python
def simulated_annealing(selectedNeighbour, selectedNeighbourCost, originalpermutation, originalpermutationCost):
    global T
    T*=alpha
    delta = originalpermutationCost - selectedNeighbourCost
    if delta>=0 :
        return selectedNeighbour, selectedNeighbourCost
    elif math.exp(delta/T) > random.uniform(0,1):
        return selectedNeighbour, selectedNeighbourCost
    return originalpermutation, originalpermutationCost
```

call select_neighbour and simulated_annealing

```python
def cutting_stock(requests, Length, numOfOrders):
    stocksCost = cal_cost(requests, Length)
    neighbour, neighbourCost = select_neighbour(requests, numOfOrders, Length)
    return simulated_annealing(neighbour, neighbourCost, requests, stocksCost)
    
```

main:

```python
def main(requests, StockLength,numOfOrders, acceptedAns):
    start = time.process_time()
    stocks = np.random.permutation(requests)
    permutation, cost = cutting_stock(stocks, StockLength, numOfOrders)
    i = 0
    while cost> acceptedAns:
        permutation, cost = cutting_stock(permutation, StockLength, numOfOrders)
        i+=1
    print("number of stocks:",cost)
    print("temperature", T)
    print(f'solved in {i} iterations and { time.process_time()-start } seconds')
    print(permutation)
    show(permutation,StockLength)
```
